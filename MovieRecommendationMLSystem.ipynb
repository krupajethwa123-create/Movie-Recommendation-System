```python
import pandas as pd

data = {
    "Movie": [
        "Avengers",
        "Iron Man",
        "Thor",
        "Batman",
        "Superman",
        "Joker",
        "Interstellar",
        "Inception"
    ],
    "Genre": [
        "Action SciFi Superhero",
        "Action SciFi Superhero",
        "Action Fantasy Superhero",
        "Action Crime Superhero",
        "Action SciFi Superhero",
        "Crime Drama",
        "SciFi Space Drama",
        "SciFi Thriller"
    ]
}

df = pd.DataFrame(data)
df
```

```python
#CONVERT TEXT TO NUMBERS,We use CountVectorizer
# Movie 1: "Action SciFi Superhero"
# Movie 2: "Action Crime Superhero"
# Common word list:Action, SciFi, Crime, Superhero
#Movie 1 counts:
# Action = 1
#SciFi = 1
#Crime = 0
#Superhero = 1
#[1, 1, 0, 1]
#Movie 2 counts:
#Action = 1
#SciFi = 0
#Crime = 1
#Superhero = 1
#[1, 0, 1, 1]
#These number lists are what the model compares.
###HOW THIS HELPS RECOMMENDATION
#Movies with similar word counts → similar content
#More shared words → higher similarity
#Fewer shared words → lower similarity
#So:Avengers & Iron Man → many common words → recommended
#Avengers & Joker → few common words → not recommended########
```

```python
#CONVERT TEXT TO NUMBERS
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(df["Genre"])
#Each movie = a numeric vector.Computer can compare movies
```

```python
#HOW DO WE COMPARE MOVIES?
#We use: Cosine Similarity.Measures how similar two movies are.
#Value range:
#1 → very similar
#0 → very different

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

#Each movie is compared with every other movie. Stored in a similarity matrix
#UNDERSTAND SIMILARITY MATRIX
#Example:
#Avengers vs Iron Man → HIGH similarity
#Avengers vs Joker → LOW similarity
#This matrix is the brain of the system.
```

```python
def recommend(movie_name):
    index = df[df["Movie"] == movie_name].index[0]

    similarity_scores = list(enumerate(similarity[index]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    print("Recommended Movies:")
    for i in similarity_scores[1:4]:
        print(df.iloc[i[0]]["Movie"]) #i[0] is just a row number in the table.
```

```python
recommend('Superman')
```
